```python
import numpy as np
import pandas as pd
from matplotlib.testing.widgets import mock_event

pd.set_option('display.float_format', '{:.10f}'.format)

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

from ModelPipeline import create_pipeline
import joblib

%matplotlib inline
```

```python
df_ml =  pd.read_csv('../Data/Initial Data.csv')
```

```python
X = df_ml.drop(['rowid', 'koi_disposition'], axis=1)
y = df_ml['koi_disposition']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
model_svc = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
model_pipeline = create_pipeline(model_svc)
model_pipeline.fit(X_train, y_train)
```

```python
y_pred = model_pipeline.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"SVC Modelinin Test Doğruluğu: {accuracy*100:.2f}%")
```

```python
joblib.dump(model_pipeline, '../Trained Models/SVM Model.joblib')
```
